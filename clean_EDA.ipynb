# Feature Relationship

```python
# 1) Correlation heatmap for numeric features
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ['Monthly Charges', 'Total Charges', 'Tenure Months']
corr = df[num_cols].corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation matrix (numeric features)')
plt.show()
```

```python
# 2) Pairplot (pairwise relationships colored by Churn)
# If the dataset is large, consider sampling before pairplot: df.sample(1000, random_state=1)
sns.pairplot(df[num_cols + ['Churn Label']], hue='Churn Label', diag_kind='kde', corner=True, plot_kws={'alpha':0.6})
plt.suptitle('Pairplot of numeric features by Churn', y=1.02)
plt.show()
```

```python
# 3) Key scatterplots with LOWESS/regression overlays
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Tenure Months', y='Total Charges', hue='Churn Label', alpha=0.6)
sns.regplot(data=df, x='Tenure Months', y='Total Charges', scatter=False, color='gray')
plt.title('Total Charges vs Tenure (with LOWESS)')
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Monthly Charges', y='Total Charges', hue='Churn Label', alpha=0.6)
sns.regplot(data=df, x='Monthly Charges', y='Total Charges', scatter=False, color='gray')
plt.title('Total Charges vs Monthly Charges (with LOWESS)')
plt.show()
```

```python
# 4) Groupby summaries and statistical tests
import numpy as np
from scipy import stats

print('Group summaries (count, mean, median, std):')
print(df.groupby('Churn Label')[num_cols].agg(['count','mean','median','std']).T)

print('\nStatistical tests (t-test / Mann-Whitney) between Churn groups:')
for col in num_cols:
    g_no = df[df['Churn Label']=='No'][col].dropna()
    g_yes = df[df['Churn Label']=='Yes'][col].dropna()
    if len(g_no) < 10 or len(g_yes) < 10:
        print(f'{col}: too few samples to test reliably')
        continue
    tstat, p_t = stats.ttest_ind(g_no, g_yes, equal_var=False)
    mw_stat, p_mw = stats.mannwhitneyu(g_no, g_yes, alternative='two-sided')
    print(f'{col}: t-test p={p_t:.3e}, mannwhitney p={p_mw:.3e}')
```

```python
# 5) Categorical relationship example: Online Security vs Churn (chi-square) and stacked proportion bar
from scipy import stats
ct = pd.crosstab(df['Online Security'], df['Churn Label'])
print('Contingency table (Online Security x Churn):\n', ct)
chi2, p_chi, dof, expected = stats.chi2_contingency(ct)
print(f'Chi-square test: chi2={chi2:.3f}, p={p_chi:.3e}, dof={dof}')

# stacked bar (proportion) visualization
ct_prop = ct.div(ct.sum(axis=1), axis=0)
ct_prop.plot(kind='bar', stacked=True, figsize=(8,5), colormap='Pastel1')
plt.ylabel('Proportion')
plt.title('Proportion of Churn by Online Security')
plt.legend(title='Churn Label', bbox_to_anchor=(1.02,1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Q1: Are month-to-month customers more likely to churn?
# EN: Compare churn rates by Contract.
# VN: So sÃ¡nh tá»‰ lá»‡ churn theo loáº¡i há»£p Ä‘á»“ng.
ct = pd.crosstab(df['Contract'], df['Churn Label'])
print('Contingency table:', ct)
ct_rate = pd.crosstab(df['Contract'], df['Churn Label'], normalize='index')
print('Churn rate by Contract:', ct_rate)
from scipy import stats
chi2, p, dof, exp = stats.chi2_contingency(ct)
print(f'Chi-square p-value: {p:.3e}')
ct.plot(kind='bar', stacked=True, figsize=(6,4))
plt.title('Churn counts by Contract')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Q2: Does tenure differ between churners and non-churners?
# EN: Compare Tenure Months distributions.
# VN: Kiá»ƒm tra Tenure giá»¯a nhÃ³m churn vÃ  khÃ´ng churn.
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Churn Label', y='Tenure Months', palette='Set2')
plt.title('Tenure Months by Churn Label')
plt.show()

from scipy import stats
g_no = df[df['Churn Label']=='No']['Tenure Months'].dropna()
g_yes = df[df['Churn Label']=='Yes']['Tenure Months'].dropna()
tstat, p_t = stats.ttest_ind(g_no, g_yes, equal_var=False)
mw, p_mw = stats.mannwhitneyu(g_no, g_yes, alternative='two-sided')
print(f't-test p={p_t:.3e}, mannwhitney p={p_mw:.3e}')
```

```python
# Q3: Is higher Monthly Charges associated with higher churn?
# EN: Compare Monthly Charges distribution by churn.
# VN: Monthly Charges cÃ³ khÃ¡c biá»‡t lá»›n giá»¯a hai nhÃ³m khÃ´ng?
plt.figure(figsize=(8,4))
sns.kdeplot(data=df, x='Monthly Charges', hue='Churn Label', common_norm=False)
plt.title('Monthly Charges distribution by Churn')
plt.show()
print(df.groupby('Churn Label')['Monthly Charges'].describe())
from scipy import stats
g_no = df[df['Churn Label']=='No']['Monthly Charges'].dropna()
g_yes = df[df['Churn Label']=='Yes']['Monthly Charges'].dropna()
tstat, p_t = stats.ttest_ind(g_no, g_yes, equal_var=False)
print(f't-test p={p_t:.3e}')
```

```python
# Q4: Do specific Internet Service types show higher churn?
# EN: Compare churn rates across Internet Service.
# VN: Loáº¡i dá»‹ch vá»¥ Internet nÃ o cÃ³ nguy cÆ¡ churn cao hÆ¡n?
ct = pd.crosstab(df['Internet Service'], df['Churn Label'])
print(ct)
print('Churn rate by Internet Service:', pd.crosstab(df['Internet Service'], df['Churn Label'], normalize='index'))
chi2, p, dof, exp = stats.chi2_contingency(ct)
print(f'Chi-square p-value: {p:.3e}')
ct.plot(kind='bar', stacked=True, figsize=(7,4))
plt.title('Churn by Internet Service')
plt.tight_layout()
plt.show()
```

```python
# Q5: Do add-on services reduce churn? (Online Security, Tech Support, Online Backup)
# EN/VN: Compute churn rate for customers with/without each add-on.
addons = ['Online Security','Tech Support','Online Backup']
for a in addons:
    print('---', a, '---')
    ct = pd.crosstab(df[a], df['Churn Label'])
    print(ct)
    print('Churn rate by', a, ':')
    print(pd.crosstab(df[a], df['Churn Label'], normalize='index'))
    print()
    # plot
    ct.plot(kind='bar', stacked=True, figsize=(6,3))
    plt.title(f'Churn by {a}')
    plt.tight_layout()
    plt.show()
```

```python
# Q6: Is payment method related to churn?
# EN/VN: Compare churn rate by Payment Method.
ct = pd.crosstab(df['Payment Method'], df['Churn Label'])
print(ct)
print('Churn rate by Payment Method:', pd.crosstab(df['Payment Method'], df['Churn Label'], normalize='index'))
chi2, p, dof, exp = stats.chi2_contingency(ct)
print(f'Chi-square p-value: {p:.3e}')
ct.plot(kind='bar', stacked=True, figsize=(8,4))
plt.title('Churn by Payment Method')
plt.tight_layout()
plt.show()
```

```python
# Q7: Which features are strongest predictors of churn? (simple logistic)
# EN/VN: Train a basic logistic regression and show coefficients.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
# select simple features and encode categoricals
X = df[['Monthly Charges','Total Charges','Tenure Months']].copy()
X = X.fillna(X.median())
y = (df['Churn Label']=='Yes').astype(int)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict_proba(X_test)[:,1]
print('ROC AUC:', roc_auc_score(y_test, y_pred))
coef = pd.Series(model.coef_[0], index=X.columns).sort_values(key=abs, ascending=False)
print('Feature coefficients (logistic):')
print(coef)
# Note: scale features and add more features for production models.
```

```python
# Q8: Interaction effects: churn rate by tenure and monthly charges bins
# EN/VN: Create 2D bins and heatmap of churn rate.
df['tenure_bin'] = pd.cut(df['Tenure Months'], bins=[-1,6,12,24,48,96], labels=['0-6','7-12','13-24','25-48','49+'])
df['monthly_bin'] = pd.qcut(df['Monthly Charges'], q=4, duplicates='drop')
pivot = df.groupby(['tenure_bin','monthly_bin'])['Churn Label'].apply(lambda s: (s=='Yes').mean()).unstack()
plt.figure(figsize=(8,5))
sns.heatmap(pivot, annot=True, fmt='.2f', cmap='YlOrRd')
plt.title('Churn rate by Tenure bin and Monthly Charges quartile')
plt.ylabel('Tenure bin')
plt.xlabel('Monthly Charges bin')
plt.tight_layout()
plt.show()
```

```python
# Q9: What are common reasons for churn? (if available)
# EN/VN: Analyze Churn Reason frequencies and relation to CLTV.
if 'Churn Reason' in df.columns:
    print(df['Churn Reason'].value_counts().head(20))
    # relation to CLTV
    df['cltv_bin'] = pd.qcut(df['CLTV'].fillna(0), q=4, duplicates='drop')
    print('Churn Reason by CLTV bin:')
    print(pd.crosstab(df['Clurn Reason'] if 'Clurn Reason' in df.columns else df.get('Churn Reason'), df['cltv_bin']))
else:
    print('No Churn Reason column in dataset')
```

```python
# Q10: Segment customers by CLTV vs churn risk
# EN/VN: Bin CLTV and compute churn rate per bin to prioritize retention.
if 'CLTV' in df.columns:
    df['cltv_bin'] = pd.qcut(df['CLTV'].fillna(df['CLTV'].median()), q=5, duplicates='drop')
    seg = df.groupby('cltv_bin')['Churn Label'].apply(lambda s: (s=='Yes').mean())
    print(seg)
    seg.plot(kind='bar', figsize=(8,4))
    plt.ylabel('Churn rate')
    plt.title('Churn rate by CLTV segment')
    plt.tight_layout()
    plt.show()
else:
    print('No CLTV column in dataset')
```
